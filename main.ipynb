#  Part 2 : Product Pricing

## 1. Measures of Centrality. Calculate the mean, median, and mode for the price of products.
### What's the average price point of products listed? How does this compare with the most common price point (mode)?


```python
# Calculating the mean, median, and mode for the price of products
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]  # mode() returns a Series, so take the first element

# Displaying the results
print("Mean price:", mean_price)
print("Median price:", median_price)
print("Mode price:", mode_price)
```

## 2. Measures of Dispersion:

Determine the variance, standard deviation, range, and interquartile range for product price.
How varied are the product prices? Are there any indicators of a significant spread in prices?2

```python
# Calculating descriptive statistics for the 'price' column
variance_price = df['price'].var()
std_dev_price = df['price'].std()
min_price = df['price'].min()
max_price = df['price'].max()
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = df['price'].quantile(0.75) - df['price'].quantile(0.25)

# Displaying the results
print("Variance of price:", variance_price)
print("Standard Deviation of price:", std_dev_price)
print("Minimum price:", min_price)
print("Maximum price:", max_price)
print("1st Quartile (Q1) of price:", q1)
print("3rd Quartile (Q3) of price:", q3)
print("Interquartile Range (IQR) of price:", iqr)
```

## Visualizations:

###  Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. 
###  Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing.

```python
#  Plot a histogram to visualize the distribution of product prices. 
plt.figure(figsize=(10, 6))
sns.histplot(data=df[df['price']<100]["price"], bins=20, kde=True, color="skyblue")
plt.title('Distribution of Product price')
plt.show()
```

# Part 3: Product Ratings (categorical)
## 1. Measures of Centrality: Calculate the mean, median, and mode for the rating of products.


```python
# Calculating the mean, median, and mode for the price of products
mean_stars = df['stars'].mean()
median_stars = df['stars'].median()
mode_stars = df['stars'].mode()[0]  # mode() returns a Series, so take the first element

# Displaying the results
print("Mean stars:", mean_stars)
print("Median stars:", median_stars)
print("Mode stars:", mode_stars)
```

## 2. Measures of Dispersion: Determine the variance, standard deviation, and interquartile range for product rating.


```python
# Calculating descriptive statistics for the 'stars' column
variance_stars = df['stars'].var()
std_dev_stars = df['stars'].std()
min_stars = df['stars'].min()
max_stars = df['stars'].max()
q1 = df['stars'].quantile(0.25)
q3 = df['stars'].quantile(0.75)
iqr = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

# Displaying the results
print("Variance of stars:", variance_stars)
print("Standard Deviation of stars:", std_dev_stars)
print("Minimum stars:", min_stars)
print("Maximum stars:", max_stars)
print("1st Quartile (Q1) of stars:", q1)
print("3rd Quartile (Q3) of stars:", q3)
print("Interquartile Range (IQR) of stars:", iqr)
```

## 3. Shape of the Distribution: Calculate the skewness and kurtosis for the rating column.


```python
from scipy.stats import skew, kurtosis

# Assuming the 'stars' column in your DataFrame
stars = df['stars']

# Skewness
skewness = skew(stars)

# Kurtosis
kurt = kurtosis(stars)

# Displaying the results
print("Skewness:", skewness)
print("Kurtosis:", kurt)
```

Interpretation: 
Skewness:

Value: 0.08120730776283354
Interpretation: Skewness measures the asymmetry of the distribution. A skewness value close to 0 indicates that the distribution is fairly symmetrical. Positive skewness indicates a distribution with a long right tail, while negative skewness indicates a long left tail. In this case, the skewness is very close to 0, suggesting that the data is nearly symmetrical.
Kurtosis:

Value: -1.926004642549839
Interpretation: Kurtosis measures the "tailedness" of the distribution. A kurtosis value close to 0 (after adjusting for excess kurtosis, where 0 is normal distribution) indicates a normal distribution. Positive kurtosis indicates heavier tails, while negative kurtosis indicates lighter tails (flatter peak and thinner tails). In this case, the kurtosis is negative and significantly less than 0, suggesting a distribution that is flatter than the normal distribution with lighter tails.

## 4. Visualizations: Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['stars'], bins=30, kde=True)

# Adding skewness and kurtosis to the plot
plt.axvline(df['stars'].mean(), color='r', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(df['stars'].mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(df['stars'].mean()))
plt.text(df['stars'].mean()*1.1, max_ylim*0.8, 'Skewness: {:.2f}'.format(skewness))
plt.text(df['stars'].mean()*1.1, max_ylim*0.7, 'Kurtosis: {:.2f}'.format(kurt))
```
